```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Concrete_data.csv')
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='rbf')
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
```

```python
plt.scatter(y_test,regressor.predict(X_test),color='red')
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],color='blue')
plt.title('SVR without Feature Scaling')
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
plt.text(5,75,f'R-squared: {r2_score(y_test, y_pred):.2f}',fontsize=12,color='black',bbox=dict(facecolor='white',edgecolor='gray',alpha=0.5))
plt.tight_layout()
plt.show()
plt.savefig("SVR_without_fs.png", dpi=300, bbox_inches='tight')
```
